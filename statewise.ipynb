```python
from geopy.geocoders import Nominatim

def get_coordinates(city, state):
    # Create a Nominatim geocoder object
    geolocator = Nominatim(user_agent="geoapiExercises")
    
    # Concatenate city and state names
    location_string = f"{city}, {state}"
    
    # Get location information
    location = geolocator.geocode(location_string)
    
    # Extract latitude and longitude
    if location:
        latitude = location.latitude
        longitude = location.longitude
        return latitude, longitude
    else:
        return None, None

# Example usage
city = "Thirupathi"
state = "Andhra Pradesh"
latitude, longitude = get_coordinates(city, state)
if latitude is not None and longitude is not None:
    print(f"Latitude: {latitude}, Longitude: {longitude}")
else:
    print("Coordinates not found for the given city and state.")
```

```python
import folium
from geopy.geocoders import Nominatim
def get_coordinates(city, state):
    geolocator = Nominatim(user_agent="geoapiExercises")
    location_string = f"{city}, {state}"
    location = geolocator.geocode(location_string)
    if location:
        latitude = location.latitude
        longitude = location.longitude
        return latitude, longitude
    else:
        return None, None
def plot_location_on_map(city, state):
    latitude, longitude = get_coordinates(city, state)
    m = folium.Map(location=[latitude, longitude], zoom_start=10)
    folium.Marker(location=[latitude, longitude], popup=f"{city}, {state}").add_to(m)
    display(m)
city = "Thirupathi"
state = "Andhra Pradesh"
plot_location_on_map(city, state)
```
